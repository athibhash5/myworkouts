```python
import seaborn as sns
from matplotlib import pyplot as plt
tips=sns.load_dataset('tips')
tips.head()

```

```python
#lmplot
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,fit_reg=False)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,fit_reg=False,hue='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,fit_reg=False,col='smoker')
plt.show()
```

```python
tips['smoker'].value_counts()
sns.lmplot(x='total_bill',y='tip',data=tips,hue='day',markers=['*','^','o','.'],col='smoker')
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='smoker',markers=['*','^'],col='day')
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='smoker',markers=['*','^'],col='day',row='sex')
plt.show()
```
