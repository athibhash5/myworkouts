```python
data=data.drop_duplicates()
data.shape
```

```python
x=data.iloc[:,:8]
x.head()
```

```python
y=data.iloc[:,8]
y.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
ytrain.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
sm=SMOTE()
```

SMOTE:(**synthetic minority oversampling technique**) for data balancing.
It involves creating synthetic samples for the minority class (the class with fewer instances) to balance the class distribution. This is particularly useful in classification tasks where one class is significantly underrepresented compared to the other.

```python
xtrain1,ytrain1=sm.fit_resample(xtrain,ytrain)
```

```python
ytrain1.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
sd=StandardScaler()
xtrain2=sd.fit_transform(xtrain1)
xtest1=sd.fit_transform(xtest)

```

```python
from sklearn.naive_bayes import GaussianNB
gs=GaussianNB()
gs.fit(xtrain2,ytrain1)
```

```python
ypred=gs.predict(xtest1)
ypred
```
