#Grid Search CV

Grid Search Cross-Validation (Grid Search CV) is a technique used for hyperparameter tuning in machine learning. It's a systematic way of searching through a predefined set of hyperparameter combinations, evaluating the model's performance for each combination using cross-validation, and selecting the hyperparameters that give the best performance.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV #to select best parameter
from sklearn import metrics
```

```python
data=pd.read_csv('/content/Copy of diabetes (1) (2) (1).csv')
data.head()
```

```python
x=data.drop('Outcome',axis=1)
y=data.Outcome
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=0,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
cls=KNeighborsClassifier(n_neighbors=5)
cls.fit(xtrain,ytrain)
ypred=cls.predict(xtest)
ypred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,ypred)
```

```python
help(cls)
```

```python
cls1=KNeighborsClassifier()
parameters={'n_neighbors':[3,5,7,9],'weights':['uniform','distance']}
clf=GridSearchCV(cls1,parameters,cv=10,scoring='accuracy')
```

The GridSearchCV object (clf) will search through these hyperparameter combinations, performing 10-fold cross-validation and optimizing for accuracy. Finally, it will print the best hyperparameters and evaluate the best model on the test set.

```python
clf.fit(xtrain,ytrain)
clf.best_params_
```

```python
cls2=KNeighborsClassifier(n_neighbors=9)
cls2.fit(xtrain,ytrain)
ypred1=cls2.predict(xtest)
ypred1
```

```python
accuracy_score(ytest,ypred1)
```
